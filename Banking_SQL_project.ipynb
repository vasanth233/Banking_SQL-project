1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

```python
# Q1: Customer ID, Customer Name, Average Balance per Customer
query = """
SELECT
    c.customer_id,
    c.customer_name,
    AVG(a.Balance_amount) AS average_balance
FROM BANK_CUSTOMER c
LEFT JOIN BANK_ACCOUNT a
    ON c.customer_id = a.Customer_id
GROUP BY c.customer_id, c.customer_name
ORDER BY c.customer_id
LIMIT 8;
"""

cursor.execute(query)

# Fetch column names
columns = [desc[0] for desc in cursor.description]

# Fetch all results
results = cursor.fetchall()

# Print table header
print("{:<12} {:<20} {:<15}".format(*columns))
print("-" * 50)

# Print each row formatted
for row in results:
    avg_bal = 0 if row[2] is None else row[2]   # Handle NULL
    print("{:<12} {:<20} {:<15.2f}".format(row[0], row[1], avg_bal))
```

7) Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)

```python
# Q7: Deducted balance for Primary (P) relationship accounts
query = """
SELECT
    a.Customer_id,
    a.Account_Number,
    a.Account_type,
    a.Balance_amount
        + COALESCE(SUM(CASE
                           WHEN t.Transaction_amount < 0
                           THEN t.Transaction_amount
                           ELSE 0
                       END), 0) AS deducted_balance_amount
FROM BANK_ACCOUNT a  -- Changed from Bank_Account_Details to BANK_ACCOUNT
LEFT JOIN BANK_ACCOUNT_TRANSACTION t
    ON a.Account_Number = t.Account_Number
WHERE a.Relation_ship = 'P' -- Changed from Relationship_type to Relation_ship
GROUP BY
    a.Customer_id,
    a.Account_Number,
    a.Account_type,
    a.Balance_amount
ORDER BY
    a.Customer_id,
    a.Account_Number;
"""

cursor.execute(query)

# Fetch column names
columns = [desc[0] for desc in cursor.description]

# Fetch all results
results = cursor.fetchall()

# Print header
print("{:<12} {:<20} {:<20} {:<22}".format(*columns))
print("-" * 80)

# Print each row
for row in results:
    cust_id = row[0]
    acc_no = row[1]
    acc_type = row[2]
    ded_bal = 0 if row[3] is None else row[3]
    print("{:<12} {:<20} {:<20} {:<22.2f}".format(cust_id, acc_no, acc_type, ded_bal))
```

10) Display all saving account holders have Add-on Credit Cards" and Credit cards" (3 Rows)

```python
# Q10: Customers who have Savings + Credit Card + Add-on Credit Card
query = """
SELECT DISTINCT
    c.customer_id,
    c.customer_name,
    a.Account_Number AS savings_account,
    cc.Account_Number AS credit_card,
    acc.Account_Number AS addon_credit_card
FROM BANK_CUSTOMER c

-- Savings accounts
JOIN Bank_Account_Details a
    ON c.customer_id = a.Customer_id
    AND a.Account_type = 'SAVINGS'

-- Credit Card accounts
JOIN Bank_Account_Details cc
    ON c.customer_id = cc.Customer_id
    AND cc.Account_type = 'Credit Card'

-- Add-on Credit Card accounts
JOIN Bank_Account_Details acc
    ON c.customer_id = acc.Customer_id
    AND acc.Account_type = 'Add-on Credit Card'

ORDER BY c.customer_id;
"""

cursor.execute(query)

# Fetch columns
columns = [desc[0] for desc in cursor.description]

# Print header
print("{:<12} {:<15} {:<20} {:<20} {:<25}".format(*columns))
print("-" * 95)

# Print rows
for row in cursor.fetchall():
    print("{:<12} {:<15} {:<20} {:<20} {:<25}".format(*row))
```

```python
# Q11: SAVINGS accounts linked with Credit Cards and aggregate sum of their transaction amount
query = """
SELECT
    s.Customer_id,
    s.Account_Number      AS savings_account,
    cc.Account_Number     AS credit_card_account,
    SUM(t.Transaction_amount) AS credit_aggregate_sum
FROM Bank_Account_Relationship_Details s
JOIN Bank_Account_Relationship_Details cc
    ON cc.Linking_Account_Number = s.Account_Number
   AND cc.Account_type = 'Credit Card'
LEFT JOIN BANK_ACCOUNT_TRANSACTION t
    ON t.Account_Number = cc.Account_Number
WHERE s.Account_type = 'SAVINGS'
GROUP BY
    s.Customer_id,
    s.Account_Number,
    cc.Account_Number;
"""

cursor.execute(query)

# Fetch columns
columns = [desc[0] for desc in cursor.description]

# Print header
print("{:<12} {:<20} {:<22} {:<25}".format(*columns))
print("-" * 85)

# Print rows
for row in cursor.fetchall():
    cust_id = row[0]
    sav_acc = row[1]
    cc_acc = row[2]
    agg_sum = 0 if row[3] is None else row[3]
    print("{:<12} {:<20} {:<22} {:<25.2f}".format(cust_id, sav_acc, cc_acc, agg_sum))
```
